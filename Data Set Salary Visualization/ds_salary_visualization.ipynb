```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("ds_salaries.csv")

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Plot 1: Histogram of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], bins=20, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Bar plot of average salaries based on experience level
avg_salary_by_exp = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_exp.index, y=avg_salary_by_exp.values)
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()

# Plot 3: Scatter plot of years of work experience vs. salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='work_year', y='salary_in_usd', data=df)
plt.title('Years of Work Experience vs. Salary')
plt.xlabel('Years of Work Experience')
plt.ylabel('Salary (USD)')
plt.show()

```
